```python
import pandas as pd
```

```python
day1_prices = pd.read_csv('../../data/round1/prices_round_1_day_-2.csv', sep=';')
day1_prices.head()
```

```python
day2_prices = pd.read_csv('../../data/round1/prices_round_1_day_-1.csv', sep=';')
day2_prices.head()


```

```python
day3_prices = pd.read_csv('../../data/round1/prices_round_1_day_0.csv', sep=';')
day3_prices.head()
```

```python
# Concatenate all days into one dataframe
all_prices = pd.concat([day1_prices, day2_prices, day3_prices], axis=0)
all_prices = all_prices.reset_index(drop=True)
all_prices.head()
```

```python
rr_prices = all_prices[all_prices['product'] == 'RAINFOREST_RESIN']
rr_prices.head()

# Calculate the average price of the RAINFOREST_RESIN product
d1_price = rr_prices['mid_price'][rr_prices['day'] == -2].mean()
d2_price = rr_prices['mid_price'][rr_prices['day'] == -1].mean()
d3_price = rr_prices['mid_price'][rr_prices['day'] == 0].mean()

print(d1_price, d2_price, d3_price)

# Calculate the average price of the RAINFOREST_RESIN product

```
