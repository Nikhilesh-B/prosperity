```python
import pandas as pd
import numpy as np
```

```python
day1_prices = pd.read_csv('../../data/round1/prices_round_1_day_-2.csv', sep=';')
day1_prices.head()
```

```python
day2_prices = pd.read_csv('../../data/round1/prices_round_1_day_-1.csv', sep=';')
day1_prices['day'] = -1
day2_prices.head()


```

```python
day3_prices = pd.read_csv('../../data/round1/prices_round_1_day_0.csv', sep=';')
day1_prices['day'] = 0
day3_prices.head()
```

```python
# Concatenate all days into one dataframe
all_prices = pd.concat([day1_prices, day2_prices, day3_prices], axis=0)
all_prices = all_prices.reset_index(drop=True)
all_prices.head()
```

```python
kelp_prices = all_prices[all_prices['product'] == 'KELP']
kelp_prices.head()

# Calculate the average price of the RAINFOREST_RESIN product
d1_price = kelp_prices['mid_price'][kelp_prices['day'] == -2].mean()
d2_price = kelp_prices['mid_price'][kelp_prices['day'] == -1].mean()
d3_price = kelp_prices['mid_price'][kelp_prices['day'] == 0].mean()

print(d1_price, d2_price, d3_price)

# Calculate the average price of the RAINFOREST_RESIN product

```

```python
# Calculate the average price of the RAINFOREST_RESIN product
d1_std = kelp_prices['mid_price'][kelp_prices['day'] == -2].std()
d1_std = kelp_prices['mid_price'][kelp_prices['day'] == -1].std()
d3_std = kelp_prices['mid_price'][kelp_prices['day'] == 0].std()


print(d1_std, d1_std, d3_std)
```

```python
# Create average bid price column considering all available bid prices
all_prices['avg_bid'] = all_prices.apply(
    lambda row: np.nanmean([
        row['bid_price_1'] if not pd.isna(row['bid_price_1']) else np.nan,
        row['bid_price_2'] if not pd.isna(row['bid_price_2']) else np.nan, 
        row['bid_price_3'] if not pd.isna(row['bid_price_3']) else np.nan
    ]), axis=1
)

# Create average ask price column considering all available ask prices 
all_prices['avg_ask'] = all_prices.apply(
    lambda row: np.nanmean([
        row['ask_price_1'] if not pd.isna(row['ask_price_1']) else np.nan,
        row['ask_price_2'] if not pd.isna(row['ask_price_2']) else np.nan,
        row['ask_price_3'] if not pd.isna(row['ask_price_3']) else np.nan
    ]), axis=1
)

all_prices.head()


```

```python
day1_trades = pd.read_csv('../../data/round1/trades_round_1_day_-2.csv', sep=';')
day1_trades['day'] = -2
day2_trades = pd.read_csv('../../data/round1/trades_round_1_day_-1.csv', sep=';')
day2_trades['day'] = -1
day3_trades = pd.read_csv('../../data/round1/trades_round_1_day_0.csv', sep=';')
day3_trades['day'] = 0

trades = pd.concat([day1_trades, day2_trades, day3_trades], axis=0)
trades = trades.reset_index(drop=True)
trades.head()
```

```python
# Create a new table that aggregates the trades table and calculates the weighted price for trades of the same timestamp and symbol
trades['weighted_price'] = trades['price'] * trades['quantity']
trades.head()

aggregated_trades = trades.groupby(['timestamp', 'symbol', 'day']).agg(
    total_quantity=('quantity', 'sum'),
    weighted_price_sum=('weighted_price', 'sum')
).reset_index()

aggregated_trades.head()

aggregated_trades['weighted_price'] = aggregated_trades['weighted_price_sum'] / aggregated_trades['total_quantity']
aggregated_trades = aggregated_trades.drop(columns=['weighted_price_sum'])
aggregated_trades = aggregated_trades[['timestamp', 'day', 'symbol', 'total_quantity', 'weighted_price']]


aggregated_trades
```

```python
# Fill in missing timestamps with the last known weighted price for the corresponding symbol
all_timestamps = pd.DataFrame({'timestamp': range(0, trades['timestamp'].max() + 100, 100)})
all_symbols = aggregated_trades['symbol'].unique()
all_days = aggregated_trades['day'].unique()

filled_aggregated_trades = pd.DataFrame()

for symbol in all_symbols:
    for day in all_days:
        symbol_day_trades = aggregated_trades[(aggregated_trades['symbol'] == symbol) & (aggregated_trades['day'] == day)]
        symbol_day_trades = all_timestamps.merge(symbol_day_trades, on='timestamp', how='left')
        symbol_day_trades['symbol'] = symbol
        symbol_day_trades['day'] = day
        symbol_day_trades['weighted_price'] = symbol_day_trades.groupby('symbol')['weighted_price'].ffill()
        filled_aggregated_trades = pd.concat([filled_aggregated_trades, symbol_day_trades], axis=0)

filled_aggregated_trades = filled_aggregated_trades.reset_index(drop=True)
filled_aggregated_trades.head()

filled_aggregated_trades
```

```python
# Create a new column 'bid_increase_indicator' to indicate if avg_bid increased from the previous timestamp for the same symbol
all_prices['bid_increase_indicator'] = all_prices.groupby('product')['avg_bid'].diff().apply(lambda x: 1 if x > 0 else 0)


all_prices
```

```python
# Create a new column 'weighted_price_increase_indicator' to indicate if weighted_price increased from the previous timestamp for the same symbol
filled_aggregated_trades['weighted_price_increase_indicator'] = filled_aggregated_trades.groupby('symbol')['weighted_price'].diff().apply(lambda x: 1 if x > 0 else 0)

filled_aggregated_trades
```

```python
# Merge the all_prices and filled_aggregated_trades dataframes on timestamp and symbol
merged_data = pd.merge(all_prices, filled_aggregated_trades, left_on=['timestamp', 'product'], right_on=['timestamp', 'symbol'], how='inner')

# Shift the weighted_price_increase_indicator by one time period to align with the bid_increase_indicator
merged_data['next_weighted_price_increase'] = merged_data.groupby('symbol')['weighted_price_increase_indicator'].shift(-1)

# Calculate the probability of weighted price increase given a bid increase in the prior time period
probability = merged_data[merged_data['bid_increase_indicator'] == 1]['next_weighted_price_increase'].mean()

print(f"The probability that the weighted price increases given that the bid increased in the prior time period is: {probability:.2f}")

merged_data.head()
```

```python

# First, let's check if the 'price_decrease_indicator' column exists in the filled_aggregated_trades dataframe
if 'price_decrease_indicator' not in filled_aggregated_trades.columns:
    # If the column does not exist, create it
    filled_aggregated_trades['price_decrease_indicator'] = filled_aggregated_trades.groupby('symbol')['weighted_price'].diff().apply(lambda x: 1 if x < 0 else 0)

# Now, let's check if the merge operation included the 'price_decrease_indicator' column
if 'price_decrease_indicator' not in merged_data.columns:
    # If the column does not exist in the merged dataframe, merge the dataframes again to include the new column
    merged_data = pd.merge(all_prices, filled_aggregated_trades, left_on=['timestamp', 'product'], right_on=['timestamp', 'symbol'], how='inner')

# Shift the price_decrease_indicator by one time period to align with the bid_increase_indicator
merged_data['next_price_decrease'] = merged_data.groupby('symbol')['price_decrease_indicator'].shift(-1)

# Calculate the probability of transacted price decrease given a bid increase in the prior time period
probability_decrease = merged_data[merged_data['bid_increase_indicator'] == 1]['next_price_decrease'].mean()

print(f"The probability that the transacted price decreases given that the bid increased in the prior time period is: {probability_decrease:.2f}")

```
